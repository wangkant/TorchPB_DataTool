```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/sample_raw.csv")  # default: sample raw data from scripts/generate_sample_data.py; replace with your own raw experiment CSV

df.sample(5)
```

```python
df["Mol Name"] = df["file"].str.extract(r"/([^/]+)_g\d{2,4}")
```

```python
print(df[["file", "Mol Name"]].head())
```

```python
df.head()
```

```python
df["Solver Name"] = df["file"].str.extract(r"/solver_(\d)/")
```

```python
df.head()
```

```python
df = df.rename(columns={
    "system": "Split",
    "NATOM": "Atom Number",
    "grid": "Grid Size",
    "EPB": "EPB Value",
    "Total_time": "Elapsed Time",
    "init_lr": "Initial Learning Rate",
    "final_lr": "Final Learning Rate",
    "CUDA_used": "CUDA Used",
    "Run_Done": "Succeeded",
    "accept": "Tolerance"
})
```

```python
df.head()
```

```python
df[df["Elapsed Time"].isna()]
```

```python
df.head()
```

```python
df[df["Elapsed Time"].isna()]
```

```python
df[df["EPB Value"].isna()]
```

```python
df["Succeeded"] = (~df["EPB Value"].isna()).astype(int)
```

```python
df["Split"] = df["Split"].replace({
    "nucleic_acid_test": "Nucleic Acid",
    "protein_test": "Protein",
    "protein_complex_test": "Protein Complex",
    "small_molecule_test": "Small Molecule"
})
```

```python
df.head(50000)
```

```python
df["Solver Name"] = df["Solver Name"].replace({
    "0": "TorchPBCGSolver",
    "1": "TorchPBBiCGSolver",
    "4": "TorchPBAdamSolver",
    "6": "TorchPBSorSolver",
})
```

```python
df.head()
```

```python
df = df[["Split", "Mol Name", "Atom Number", "Grid Size", "Solver Name", "Tolerance", "EPB Value", "Elapsed Time", "CUDA Used", "Succeeded", "Initial Learning Rate", "Final Learning Rate"]]
```

```python
df.head()
```

```python
df["Mol Name"] = df["Mol Name"].str.replace(".p22", "", regex=False)
```

```python
df["Elapsed Time"] = df["Elapsed Time"].round(2)
```

```python
df["EPB Value"] = df["EPB Value"].round(4)
```

```python
df["Sor Omega"] = pd.NA
```

```python
df["Other Comments"] = pd.NA
```

```python
sm_cg = df[df["Solver Name"] == "TorchPBCGSolver"]
sm_bicg = df[df["Solver Name"] == "TorchPBBiCGSolver"]
sm_sor = df[df["Solver Name"] == "TorchPBSorSolver"]

display(sm_cg)
display(sm_bicg)
display(sm_sor)
```

```python
# default output path for the cleaned SOR-solver slice; change as needed
sm_sor.to_csv("data/cleaned_sor.csv", index=False, encoding="utf-8-sig")
```
